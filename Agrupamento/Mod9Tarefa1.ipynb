# Tarefa 1

### 1. Idade media das mães e dos pais por município (coluna munResNome)


```python
sinasc_select = sinasc_raw[['IDADEMAE', 'IDADEPAI', 'munResNome']].dropna()
result = sinasc_select.groupby('munResNome').mean()
result
```

```python
Descobri que na grande maioria das vezes a idade média dos pais é maior que a idade média das mães.
```

### 2. Peso médio dos bebes por sexo que nasceram no dia do seu aniversário por faixas de escolaridade mae
Ex: Você, aluna(o), nasceu no dia 10/01, então você precisa filtrar o conjunto de dados nessa data e calcular o peso médio dos bebês de cada sexo por faixa de escolaridade da mãe.

```python
sinasc_aniv = sinasc_raw[['ESCMAE', 'DTNASC', 'PESO', 'SEXO']].dropna()
data_aniv = sinasc_aniv[sinasc_aniv['DTNASC'] == '2019-08-02']
result = data_aniv.groupby(['ESCMAE', 'SEXO']).PESO.mean()
result.to_frame()
```

```python
Descobri que crianças que mãe tem tem um grau de escolaridade maior nascem mais pesadas.
```

### 3. Qual o municipio que nasceu menos bebe em 2019?
    - qual a idade media, maxima, minima das maes nesse municipio?
    - qual a idade media, maxima, minima dos pais nesse municipio?

```python
sinasc_mun = sinasc_raw[['IDADEMAE', 'IDADEPAI', 'munResNome']]
result = sinasc_select.groupby('munResNome')['munResNome'].count().sort_values()
result.to_frame()
```

```python
result2 = sinasc_mun.groupby('munResNome').aggregate(['min', 'mean', 'max'])
result2
```

### 4. Qual o municipio que nasceu mais bebe no mês de março?
    - qual a quantidade de filhos vivos media, maxima, minima nesse municipio?
    - qual a idade media, maxima, minima dos pais nesse municipio?



```python
sinasc_select = sinasc_raw[['IDADEMAE', 'IDADEPAI', 'munResNome', 'QTDFILVIVO']].dropna()
munMaisNasc = sinasc_select[sinasc_select['munResNome'] == 'Porto Velho']
result1 = munMaisNasc.groupby('munResNome').agg(['mean', 'max', 'min'])
result1
```
