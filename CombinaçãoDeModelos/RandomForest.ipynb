## Bootstrao + Feature selection:

Bootstrap é a seleção randomica das linhas do dataframe, e o Feature selectio é quando escolhemos uma quantidade de colunas e elas serão selecionadas de maneira randomica.  

A quantidade de colunas é esolhida não por acaso. Segundo o criador Leo Breimer devemos utilizar como padrão as seguintes condições:  

Para problemas de classificação: -> o numero de colunas da amostra é igual a raiz do numero de colunas do dataframe.  

Para problemas de regressão -> o numero de colunas da amostra é igual ao numero de colunas do dataframe dividido por 3.  

Em seguida para cada amostra é criada uma arvore de forma aleatória que serão classificadas, vistas a frequencia de cada classificação e o resultado é a classificação mais frequente.

Há uma grande diferença entre Bagging e Random Foreste. Uma delas é o tipo de modelagem usado, outra diferença é a seleção das colunas.

```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score
```

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
titanic.info()
```

```python
titanic = titanic.drop(columns=['deck', 'alive'])
titanic = titanic.dropna()
```

```python
titanic.info()
```

```python
titanic = pd.get_dummies(titanic, drop_first=True).astype(int)
titanic.head()
```

```python
y = titanic.survived
X = titanic.drop(columns='survived')
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
modelo = RandomForestClassifier(n_estimators=50, random_state=42)

modelo.fit(X_train, y_train)

y_pred = modelo.predict(X_test)
```

```python
accuracy_score(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```
