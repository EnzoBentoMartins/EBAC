```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

df = pd.read_csv('dadosSus.csv')
df
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
df_select = df[['APGAR1', 'APGAR5', 'SEXO', 'GESTACAO', 'ESTCIVMAE', 'CONSULTAS']].copy()
df_select.head().plot.bar()
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
sns.barplot(x='CONSULTAS', y='APGAR5', estimator='mean', data=df_select)
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
fig, ax = plt.subplots(2, 2, figsize=(10, 8))

sns.countplot(ax = ax[0, 0], x='SEXO', data=df_select)
sns.countplot(ax = ax[0, 1], x='GESTACAO', data=df_select)
sns.countplot(ax = ax[1, 0], x='ESTCIVMAE', data=df_select)
sns.countplot(ax = ax[1, 1], x='CONSULTAS', data=df_select)

labels = ax[0, 1].tick_params(axis='x', rotation=15)
```
