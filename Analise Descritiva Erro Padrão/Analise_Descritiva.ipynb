1) Com o que você viu em aula, faça uma análise das variáveis qualitativas (posse de imóvel e posse de veículo) para verificar se são boas preditoras de *renda*. Alguma delas parece discriminar *renda* de forma estatisticamente significante?

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

prev_renda = pd.read_csv('previsao_de_renda.csv')
desc_renda = prev_renda[['posse_de_veiculo', 'posse_de_imovel', 'renda', 'idade']]
desc_renda.head()
````

````python
sns.pointplot(y=desc_renda['renda'], x=desc_renda['posse_de_veiculo'])
````

````python
sns.pointplot(y=desc_renda['renda'], x=desc_renda['posse_de_imovel'])
````

````python
plt.figure(figsize=(20,10))
sns.pointplot(y=desc_renda['renda'], x=desc_renda['idade'], hue=desc_renda['posse_de_veiculo'])
````

````python
plt.figure(figsize=(20,10))
sns.pointplot(y=desc_renda['renda'], x=desc_renda['idade'], hue=desc_renda['posse_de_imovel'])
````

### De fato os dois paramentros (posse de imovel e posse de veiculo) discriminam renda de forma significante. pessoas em posse de veiculos estão altamente relacionados com uma renda maior, levando em conta sua idade.
### Embora pessoas com maior renda tem mais probabilidade de posse de imoveis ainda existe um certo balanceamento, isso levanto em conta a idade.
