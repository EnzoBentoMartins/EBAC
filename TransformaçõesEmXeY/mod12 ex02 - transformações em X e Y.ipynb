# Regressão 01 - tarefa 02 - transformações em X e Y

````python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf
from sklearn.metrics import r2_score
````

### 1. Melhorando o ajuste no percentual de gorjetas

a. Vamos partir do modelo feito na última tarefa, o que relaciona ```tip_pct``` e ```net_bill```. Carregue a base, os pacotes e reproduza este modelo aqui.

````python
tips = sns.load_dataset('tips')
tips['tip_pct'] = tips['tip'] / (tips['total_bill'] - tips['tip'])
tips['net_bill'] = tips['total_bill'] - tips['tip']
tips.head()
````

````python
sns.regplot(y='tip_pct', x='net_bill', data=tips[tips['tip_pct']<.5]) 
````

````python
reg = smf.ols('tip_pct ~ net_bill', tips).fit()
print(reg.summary())

tips['res'] = reg.resid

sns.scatterplot(x='net_bill', y='res', data=tips, alpha=.75)
plt.axhline(y=0, color='r', linestyle='--')
````

### 2. ajuste o modelo de ```tip_pct``` em função do logaritmo de ```net_bill```.

````python
reg = smf.ols('tip_pct ~ np.log(net_bill)', tips).fit()
print(reg.summary())

tips['res'] = reg.resid

sns.scatterplot(x='net_bill', y='res', data=tips, alpha=.75)
plt.axhline(y=0, color='r', linestyle='--')
````

###  3. ajuste o modelo de ```tip_pct``` em função de um polinômio de segundo grau em ```net_bill```.

````python
reg = smf.ols('tip_pct ~ net_bill + np.power(net_bill,2)', tips).fit()
print(reg.summary())

tips['res'] = reg.resid

sns.scatterplot(x='net_bill', y='res', data=tips, alpha=.75)
plt.axhline(y=0, color='r', linestyle='--')
````

###  4. ajuste um modelo no log do ```tip_pct``` em função de ```net_bill```. Calcule o $R^2$ também em ```tip_pct```

````python
reg = smf.ols('np.log(tip_pct) ~ net_bill', tips).fit()
print(reg.summary())

tips['res'] = reg.resid

sns.scatterplot(x='net_bill', y='res', data=tips, alpha=.75)
plt.axhline(y=0, color='r', linestyle='--')

tips['pred'] = np.exp(reg.fittedvalues)
tips[['pred', 'tip_pct']].corr().iloc[0,1]**2
````

###  5. ajuste um modelo no log do ```tip_pct``` em função do log de ```net_bill```. Calcule o $R^2$ também em ```tip_pct```.

````python
reg = smf.ols('np.log(tip_pct) ~ np.log(net_bill)', tips).fit()
print(reg.summary())

tips['res'] = reg.resid

sns.scatterplot(x='net_bill', y='res', data=tips, alpha=.75)
plt.axhline(y=0, color='r', linestyle='--')

tips['pred'] = np.exp(reg.fittedvalues)
tips[['pred', 'tip_pct']].corr().iloc[0,1]**2
````

###  6. qual modelo resultou no melhor ajuste?

O ultimo modelo foi o que melhor performou.  
Embora o modelo relacionando tip_pct e net_bill não seja tão performatico. Tanto pela sua performace explicativa baixa quanto por ter um outlier significativo, o ultimo modelo usando o log tanto em x quanto em y foi o que melhor performou tendo um R² de 30,8%.
