# Gradient Boosting Machine - GBM

Para realisar o GBM, primeiro passo é tirar a média dos valores de Y, a primeira arvore vem a partir dessa média. Segundo passo é calcular o residuo (diferença entre o Y real e o Y predito). A arvore vai ser calculada atraves desse residuo, formando um novo predito. para evitar overfitting, ao inves de somar o primeiro predito com o segundo, é somado o primeiro predito com o eta vezes o segundo predito.

Tres principais diferenças entre GBM e AdaBoost:  
I. AdaBoost é uma floresta de stamps, euqnato GBM é mais semelhante com Random Forest por ser uma floresta de arvores.  
II. enquanto no AdaBosst o primeiro passo é um stamp, enquanto no GBM o primeiro passo é a média do Y.  
III. enquanto no Adaboost cada resposta tem um peso diferente, no GBM todas as respostas das arvores possuem um multiplicador em comum chamado learning_rate (eta).  
IV. AdaBoost é mais sensivel a outliers, pois aumenta o peso de respostas mal classificadas, enquanto GBM tende a ser mais robusto a outliers, especialmente se técnicas como regularização ou subsampling forem utilizadas.  
V. AdaBoost é mais propenso a overfitar, ainda mais se o numero de classificadores fracos forem muito altos, equanto GBM possui várias técnicas de controle de overfitting, como a regularização e o ajuste da profundidade das árvores.

```python
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.datasets import load_iris
```

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
titanic.info()
```

```python
titanic = titanic.drop(columns=['alive', 'deck'])
titanic.info()
```

```python
titanic = titanic.dropna()
titanic = pd.get_dummies(titanic, drop_first=True).astype(int)
titanic.info()
```

```python
y = titanic.survived
X = titanic.drop(columns='survived')
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=45)
```

```python
modelo = GradientBoostingClassifier(n_estimators=100, learning_rate=0.1, max_depth=3, random_state=45)
modelo.fit(X_train, y_train)
```

```python
y_pred = modelo.predict(X_test)
```

```python
acuracia = accuracy_score(y_test, y_pred)
print(acuracia)
```

```python
new_frame = {'y_test': y_test, 'y_pred': y_pred}
new_frame = pd.DataFrame(new_frame)
new_frame
```

```python
iris = load_iris()
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['species'] = iris.target
iris_df
```

```python
y0 = iris_df.species
X0 = iris_df.drop(columns='species')

X0_train, X0_test, y0_train, y0_test = train_test_split(X0, y0, random_state=45)
```

```python
modelo = GradientBoostingClassifier(n_estimators=50, learning_rate=0.1, max_depth=3, random_state=45)
modelo.fit(X0_train, y0_train)
```

```python
y0_pred = modelo.predict(X0_test)
```

```python
acuracia = accuracy_score(y0_test, y0_pred)
print(acuracia)
```

```python
new_frame = {'y_test': y0_test, 'y_pred': y0_pred}
new_frame = pd.DataFrame(new_frame)
new_frame
```

# Hyperparametros

I. n_estimators: O número de árvores a serem construídas.  
II. learning_rate: A taxa de aprendizado, que controla a contribuição de cada árvore para o modelo final.  
III. max_depth: A profundidade máxima de cada árvore.  
IV. min_samples_split: O número mínimo de amostras necessárias para dividir um nó.  
V. subsample: A fração das amostras a serem utilizadas para treinar cada árvore.

Conforme descrito por Jerome Friedman, a maior diferença entre os dois algoritmos está na maneira como os dados são utilizados para treinar cada árvore no ensemble.
