```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

pd.set_option('display.max_columns', None)
df = pd.read_csv('dadosSus.csv')
df.head()
```

#### 2) Categorize a variável APGAR5 conforme as seguintes quebras:

- 0 a 3: 'asfixia grave'
- 4 a 5: 'asfixia moderada'
- 6 a 7: 'asfixia leve'
- 8 a 10: 'normal'

```python
df.loc[(df['APGAR5'] >= 0) & (df['APGAR5'] <= 3), 'APGAR5CAT'] = 'Asfixia grave'
df.loc[(df['APGAR5'] >= 3) & (df['APGAR5'] <= 5), 'APGAR5CAT'] = 'Asfixia moderada'
df.loc[(df['APGAR5'] >= 6) & (df['APGAR5'] <= 7), 'APGAR5CAT'] = 'Asfixia leve'
df.loc[df['APGAR5'] >= 8, 'APGAR5CAT'] = 'Normal'
df.tail(10)
```

#### 3) Monte um gráfico comparando a distribuição de APGAR5 por sexo para cada grupo de tempo de gestação (GESTACAO).

```python
sns.catplot(
    x='GESTACAO',
    y='APGAR5',
    hue='SEXO',
    kind='bar',
    estimator=sum,
    data=df)

xlabel = plt.xticks(rotation=90)
```

####  4) Monte uma tabela cruzada com APGAR5 nas colunas e GESTACAO nas linhas

```python
tc = pd.crosstab(df['GESTACAO'], df['APGAR5'])
tc
```

####  5) Da tabela cruzada do intem anterior, calcule os percentuais de linha. Ou seja, a distribuição da variável APGAR5 para cada classe de GESTACAO dado.

```python
tc_pct = tc.div(tc.sum(axis=1), axis=0)
tc_pct
```
